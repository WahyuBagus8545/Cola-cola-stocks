```python
import pandas as pd
import numpy as np
```

```python
# Load the CSV file
df = pd.read_csv('/content/coco-cola-stock-data-19622021/COCO COLA.csv')

print(df)
```

```python
# Identify and drop the unnamed column
Unnamed_column = df.columns[df.columns.str.contains('Unnamed')].tolist()

# Drop the unnamed column if it exists
if Unnamed_column:
    df = df.drop(columns=Unnamed_column)
```

```python
import matplotlib.pyplot as plt

dates = df['Date'].values
close = df['Close'].values

plt.figure(figsize=(15,5))
plt.plot(dates, close)
plt.title('graf closingan',
          fontsize=20)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Reshape the data to make it a 2D array
close_reshaped = close.reshape(-1, 1)

# Create a MinMaxScaler and apply fit_transform
scaler = MinMaxScaler()
close_scaled = scaler.fit_transform(close_reshaped)
```

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf

X_train, X_val, y_train, y_val = train_test_split(dates, close_scaled, test_size=0.2, random_state=42,shuffle=False)
```

```python
def windowed_dataset(series, window_size, batch_size, shuffle_buffer):
    series = tf.expand_dims(series, axis=-1)
    ds = tf.data.Dataset.from_tensor_slices(series)
    ds = ds.window(window_size + 1, shift=1, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(window_size + 1))
    ds = ds.shuffle(shuffle_buffer)
    ds = ds.map(lambda w: (w[:-1], w[-1:]))
    return ds.batch(batch_size).prefetch(1)
```

```python
train_set = windowed_dataset(y_train, window_size=60, batch_size=100, shuffle_buffer=1000)
val_set = windowed_dataset(y_val, window_size=60, batch_size=100, shuffle_buffer=1000)
```

```python
# Check the shape of the first batch in the training set
for batch in train_set.take(1):
    inputs, targets = batch
    print("Input shape:", inputs.shape)
    print("Target shape:", targets.shape)

for batch in val_set.take(1):
    inputs, targets = batch
    print("Input shape:", inputs.shape)
    print("Target shape:", targets.shape)
```

```python
from keras.layers import Dense, LSTM, Dropout
from keras.models import Sequential

input_shape = (60, 1)  # Adjust the input shape based on your data
target_shape = (1, 1, 1)  # Adjust the target shape based on your data

model = Sequential([
    LSTM(60, return_sequences=True, input_shape=input_shape),
    Dropout(0.2),  # Add dropout layer
    LSTM(60),
    Dropout(0.2),  # Add dropout layer
    Dense(30, activation="relu"),
    Dropout(0.2),  # Add dropout layer
    Dense(10, activation="relu"),
    Dense(target_shape[-1])  # Output layer without activation for regression
])

# Build the model explicitly
model.build(input_shape=(None,) + input_shape)  # None for batch size
model.summary()
```

```python
threshold_mae = ( close_scaled.max() -  close_scaled.min()) * 10/100
print("Batas nilai 10% mae dari data adalah sebesar", threshold_mae)
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('mae') < threshold_mae):
      print("MAE < 10%")
      self.model.stop_training = True
callbacks = myCallback()
```

```python
from keras.optimizers import SGD
from keras.losses import Huber

# Set up SGD optimizer with learning rate and momentum
optimizer = SGD(learning_rate=1.0000e-04, momentum=0.9)

# Compile the model
model.compile(loss=Huber(),
              optimizer=optimizer,
              metrics=["mae"])


history = model.fit(train_set, epochs=100, validation_data=val_set, callbacks=[callbacks])

```

```python
# Assuming 'history' is the object returned by model.fit
train_loss = history.history['loss']
val_loss = history.history['val_loss']
val_mae = history.history['val_mae']
mae = history.history['mae']

epochs = range(1, len(train_loss) + 1)

# Plotting training loss
plt.plot(epochs, train_loss, label='Training Loss')
# Plotting validation loss
plt.plot(epochs, val_loss, label='Validation Loss')
# Plotting validation MAE
plt.plot(epochs, mae, label='MAE')
# Plotting validation MAE
plt.plot(epochs, val_mae, label='Validation MAE')

plt.title('Training and Validation Metrics')
plt.xlabel('Epochs')
plt.ylabel('Metrics')
plt.legend()

plt.show()
```
